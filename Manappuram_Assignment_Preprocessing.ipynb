Assignment on Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Bengaluru_House_Data.csv")
df.head()
```

Identify and treat missing values

```python
df['society'].isnull().sum()
```

Impute missing categorical values with 'Unknown'

```python
df['society'] = df['society'].fillna("Unknown")
```

Verify

```python
df['society'].value_counts().head()
```

Check for outliers in the price column

Common methods:

IQR method: Outliers are values outside [ ð‘„ 1 âˆ’ 1.5 â‹… ð¼ ð‘„ ð‘… , ð‘„ 3 + 1.5 â‹… ð¼ ð‘„ ð‘… ] .

Z-score method: Outliers are values with âˆ£ ð‘§ âˆ£ > 3

IQR is more robust for skewed housing price data.

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
outliers
```

Dropping Rows & Final Cleaned Dataset

Reasoning:

1.Rows with extreme outliers in price should be dropped because they distort model training.


2.Rows with missing society were imputed, so no need to drop them.

```python
#Drop outliers
df_cleaned = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
#Final row count
print("Original row count:", df.shape[0])
print("Outliers dropped:", outliers.shape[0])
print("Final cleaned row count:", df_cleaned.shape[0])
```



Final Answer

Imputed Society Column: Missing values filled with "Unknown".

Outliers in Price Column: Identified using IQR method.

Outliers = rows with price < Q1 - 1.5IQR or price > Q3 + 1.5IQR.

Rows Dropped: Outlier rows only.

Final Cleaned Dataset: All rows retained except outliers.
