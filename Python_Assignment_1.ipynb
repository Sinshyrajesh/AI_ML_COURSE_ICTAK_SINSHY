1. A financial analyst stores monthly revenues in a list: [120000, 135000,128000, 142000, 150000]. Find the difference between the highest and lowest revenue without
using loops.

```python
Monthly_revenues = [120000, 135000, 128000, 142000, 150000]

difference = max(Monthly_revenues) - min(Monthly_revenues)

print("Difference between highest and lowest revenue:", difference)
```

5. You have two tuples: interest_2024 = (7.1, 7.3, 7.2) and interest_2025 = (7.4, 7.5).
Combine them and print the average of all interest rates.

```python
A_2024 = (7.1, 7.3, 7.2)
B_2025 = (7.4, 7.5)
all_interest = A_2024 + B_2025
average = sum(all_interest) / len(all_interest)
print("Average interest rate:", average)
```

9. A tuple customer = ('Kiran', 'Loan', 80000, 7.5). Calculate and print the total amount
payable after 1 year if interest is simple interest based on rate 7.5%.

13. Combine two lists customer_names = ['Asha', 'John', 'Meera'] and balances = [56000,
72000, 65000] into a formatted string such as 'Asha:56000, John:72000, Meera:65000'
without loops.

```python
customer_names = ['Asha', 'John', 'Meera']
balances = [56000, 72000, 65000]
result = ', '.join(f"{n}:{b}" for n, b in zip(customer_names, balances))
print(result)
```
